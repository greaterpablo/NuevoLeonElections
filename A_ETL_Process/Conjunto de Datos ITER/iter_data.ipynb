```python
import pandas as pd
```

```python
first_path = "../../resources/data/Relacion ID-Municipios/ID-Municipios.csv"
```

```python
id_municipality = pd.read_csv(first_path, encoding='latin-1')
id_municipality.head()
```

```python
file_path = "../../resources/data/INEGI - Censo NL 2020/conjunto_de_datos_iter_19CSV20.csv"
```

```python
iter_data = pd.read_csv(file_path)
clean_data = iter_data[["NOM_MUN","NOM_LOC","LOC","POBTOT","GRAPROES","PEA","PDESOCUP","PSINDER"]]
clean_data
```

```python
specific_df = clean_data.loc[clean_data["LOC"] == 0,:]
cleaned_df = specific_df.drop(index =0)
cleaned_df = cleaned_df.reset_index()
cleaned_df = cleaned_df[["NOM_MUN","NOM_LOC","LOC","POBTOT","GRAPROES","PEA","PDESOCUP","PSINDER"]]
cleaned_df
```

```python
ids = []
x = 19001
c = 0
while c < 51:
    ids.append(x)
    x = x + 1
    c = c + 1
```

```python
cleaned_df["ID"] = ids
```

```python
cleaned_df.head()
```

```python
ultimate_clean = cleaned_df[["ID","NOM_MUN","NOM_LOC","LOC","POBTOT","GRAPROES","PEA","PDESOCUP","PSINDER"]]
ultimate_clean.head()
```

```python
approved_df = pd.merge(ultimate_clean, id_municipality, on = "ID", how = "inner")
approved_df = approved_df[["ID","Municipio","NOM_LOC","LOC","POBTOT","GRAPROES","PEA","PDESOCUP","PSINDER"]]
```
