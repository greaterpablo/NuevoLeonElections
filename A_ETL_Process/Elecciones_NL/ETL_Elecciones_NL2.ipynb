## Ayuntamientos 2015

```python
ayuntamientos2015 = pd.read_csv("../../resources/data/CEENL - Resultados Electorales/CEENL-Ayuntamientos-2015.csv", encoding="latin-1")
print(ayuntamientos2015.shape)
ayuntamientos2015.head()
```

```python
ayuntamientos2015.iloc[:,8:].columns
```

```python
votos_a2015 = ayuntamientos2015[['Municipio','Total','LN','Partido Acción Nacional',
    'Partido Revolucionario Institucional',
    'Partido de la Revolución Democrática',
    'Partido del Trabajo',
    'Partido Verde Ecologista de México',
    'Movimiento Ciudadano',
    'Nueva Alianza Partido Político Nacional',
    'Partido Demócrata',
    'Partido Cruzada Ciudadana',
    'MORENA',
    'Partido Humanista  Partido Político Nacional',
    'Partido Político Nacional  Encuentro Social',
    'Candidato Independiente',
    'Combinación PRI PVEM NA PD',
    'Combinación PRI PVEM NA',
    'Combinación PRI PVEM PD',
    'Combinación PRI NA PD',
    'Combinación PVEM NA PD',
    'Combinación PRI PVEM',
    'Combinación PRI NA',
    'Combinación PRI PD',
    'Combinación PVEM NA',
    'Combinación PVEM PD',
    'Combinación NA PD',
    'Combinación PRD PT']].groupby('Municipio').sum()

print(votos_a2015.shape)
votos_a2015.head()
```

```python
votos_a2015.rename({'LN': 'LNominal',
                    'Partido Acción Nacional': 'PAN',
                    'Partido Revolucionario Institucional': 'PRI',
                    'Partido de la Revolución Democrática': 'PRD',
                    'Partido del Trabajo': 'PT',
                    'Partido Verde Ecologista de México': 'PVEM',
                    'Movimiento Ciudadano': 'MC',
                    'Nueva Alianza Partido Político Nacional': 'NA',
                    'Partido Demócrata': 'PD',
                    'Partido Cruzada Ciudadana': 'PCD',
                    'Partido Humanista  Partido Político Nacional': 'PH',
                    'Partido Político Nacional  Encuentro Social': 'PES',
                    'Candidato Independiente': 'IND'}, axis=1, inplace=True)
```

```python
votos_a2015.head()
```

```python
votos_a2015['PRI'] = votos_a2015['PRI'] + (votos_a2015['PVEM']*.6) + \
    votos_a2015['Combinación PRI PVEM NA PD'] + votos_a2015['Combinación PRI PVEM NA'] + \
    votos_a2015['Combinación PRI PVEM PD'] + votos_a2015['Combinación PRI NA PD'] + \
    votos_a2015['Combinación PVEM NA PD'] + votos_a2015['Combinación PRI PVEM'] + \
    votos_a2015['Combinación PRI NA'] + votos_a2015['Combinación PRI PD'] +  \
    votos_a2015['Combinación PVEM NA'] + votos_a2015['Combinación PVEM PD'] + \
    votos_a2015['Combinación NA PD'] 
votos_a2015['PRD'] = votos_a2015['PRD'] + (votos_a2015['PT']*1.01) + votos_a2015['Combinación PRD PT'] 

votos_a2015.drop('PVEM', axis=1, inplace=True)
votos_a2015.drop('PT', axis=1, inplace=True)

votos_a2015.head()
```

```python
ganadores_a2015 = votos_a2015.iloc[:,2:].idxmax(axis=1)
ganadores_a2015.to_list()[0:5]
```

```python
ganadores_a2015.value_counts()
```

```python
votos_a2015.insert(2, 'Ganador', ganadores_a2015)
print(votos_a2015.shape)
votos_a2015.head()
```

```python
resumen_a2015 = votos_a2015[['PAN', 'PRI', 'MC', 'MORENA']]
resumen_a2015.insert(4,'Otro', votos_a2015.drop(['Total', 'LNominal', 'Ganador', 'PAN', 'PRI', 'MC', 'MORENA'], axis=1, inplace=False).max(axis=1))
resumen_a2015 = resumen_a2015.astype(int)
resumen_a2015.insert(0,'Ganador', votos_a2015['Ganador'])
resumen_a2015['Total'] = votos_a2015['Total']
resumen_a2015['LNominal'] = votos_a2015['LNominal']
resumen_a2015.head()
```

```python
resultados_a2015 = ID_Municipios_2015.merge(resumen_a2015, on='Municipio')
print(resultados_a2015.shape)
resultados_a2015

## Resultados Ayuntamientos 2015:
## http://computo2015oficial.ceenl.mx/eleccionc_1_M.html
```
