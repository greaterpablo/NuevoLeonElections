```python
import pandas as pd
```

```python
coordinates_path = "../../resources/data/Coordenadas/lat_lon.csv"
```

```python
coordinates = pd.read_csv(coordinates_path, encoding='latin-1')
coordinates.head()
```

```python
first_path = "../../resources/data/Relacion ID-Municipios/ID-Municipios.csv"
```

```python
id_municipality = pd.read_csv(first_path, encoding='latin-1')
id_municipality
```

```python
approved_df = pd.merge(id_municipality, coordinates, on = "ID", how = "inner")
approved_df
```

```python
approved_df.to_csv("../output/coordinates.csv", index = True)
```

```python
approved_df.to_json("../output/coordinatesr.js", orient = "records")
```

```python
approved_df.to_json("../output/coordinatesc.js", orient = "columns")
```
